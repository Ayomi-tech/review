```python
import apache_beam as beam
```

```python
# Extracting the item fat of 'regular' from grocery txt.


p1 = beam.Pipeline()

grocery = (p1
           | "Read from Text" >> beam.io.ReadFromText("/content/sample_data/grocery.txt", skip_header_lines= 1)
           | "split the record" >> beam.Map(lambda record: record.split(',')) #split the data where theire is ','
           | "Filter regular" >> beam.Filter(lambda record: record[2]== 'Regular') #provide data with 'regular'
           | "Write to text" >> beam.io.WriteToText('regular_filter.txt')) #save the data to 'regular_filter.txt'


p1.run()
```
