# Short-circuit evaluation

```python
#Means python interpretor will stop as soon as one of the arguement is False. this is base on the logical operator.

# since high_income from the code below is False then python interpretor will stop and give it false result.

high_income = False
good_credit = True
student = True

if high_income and good_credit and not student:
    print("eligible")
else:
    print("Not_eligible")
```

```python
tags = ['yes', 'man', 'to', 'shop']
reviews = ['is the man still coming to our shop? yes, the man is on his way to the shop']


tag_counts = []

for tag in tags:
    count = 0  
    for review in reviews:
        count += review.count(tag) 
        tag_counts.append(count) 
# Print the tag counts
for i, tag in enumerate(tags):
    print(f"'{tag}'  {tag_counts[i]} ")
```

# DELIVERY HERO TASK

```python
import pandas as pd

data = {
    'order_date': ['2024-01-27', '2024-01-17', '2024-01-17', '2024-01-17', '2024-01-27'],
    'test_name': ['sample_test', 'sample_test', 'sample_test', 'sample_test', 'sample_test'],
    'test_variant': ['Variation3', 'Variation3', 'Variation7', 'Control', 'Variation4'],
    'order_code': ['o6ev-ih36', 'x0bb-g5uf', 's0oz-4w2d', 'v4hk-8kc6', 'eydq-6bci'],
    'delivery_fee': [0.98, 2.59, 2.99, 3.49, 2.79]
}

dl = pd.DataFrame(data)

print(dl)
```

```python
avg_deli_fee = dl.groupby('test_variant')['delivery_fee'].mean()
print(avg_deli_fee)
```

```python
top_2 = avg_deli_fee.nlargest(2)
second_value = top_2.iloc[1]  # Getting the second largest value using iloc
print(second_value)
```
